## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
uppercases = [x.upper() for x in words]
print(uppercases)
```

**Create a new list containing only words with 5 or more letters**

```python
more = [x for x in words if len(x) > 5]
print(more)
```

**Print the first word starting with "t"**

```python
starting = [x for x in words if x.startswith('t')]
print(starting[0])
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
square = [x ** 2 for x in range (1,11)]
print(square)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
odd_square = [x ** 2 for x in range (1,11) if x % 2 == 1]
print(odd_square)   
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
multiples_eight =[x for x in range(8,1000,8)]
print(multiples_eight)
    
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
total_names = len(people)
print(total_names)
```

**How many people have kids**?

```python
how_many = [1 for x in people if x['n_kids'] > 0]
print(len(how_many))
```

**How many kids do they have in total?**

```python
total_kids = sum([x['n_kids'] for x in people])
print(total_kids)
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
for p in people:
    if p.get("name", "").endswith("a"):
        p["n_kids"] = p.get("n_kids", 0) + 1
print(people)
```
